```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
```

```python
#Aplicação do algoritmo definindo 3 clusters e passando a matriz transposta (iris.data.T). Os outros parâmetros são obrigatórios e são os default indicados na documentação
r = skfuzzy.cmeans(data = iris.data.T, c = 3, m= 2, error = 0.005, maxiter= 1000, init = None)
```

```python
#Obtendo porcentagens de um registro pertencer a um cluster, que está na posição 1 da matriz retornada
previsoes_porcentagem = r[1]
```

```python
#Visualização da probabilidade de um registro pertencer a cada um dos cluster(o somatório é 1)
for x in range(150):
  print(previsoes_porcentagem[0][x], previsoes_porcentagem[1][x], previsoes_porcentagem[2][x])
```

```python
#Geração da matriz de contingência para a comparação com as classes originais da base de dados
previsoes = previsoes_porcentagem.argmax(axis=0)
resultado = confusion_matrix(iris.target, previsoes)
resultado
```
